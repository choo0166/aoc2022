```python
import numpy as np

data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2022/day12.txt',
    'r').readlines()

mat = []
for line in data:
    mat.append(list(line.rstrip()))

mat = np.array(mat)
mat
```

```python
import string
from collections import deque

def findShortestPath(input):
    m, n = input.shape[0], input.shape[1]
    ele = string.ascii_lowercase # a-z
    elevationMap = {symbol: height for height, symbol in enumerate(ele, start = 1)}

    def getStartingPosition(input):
        for i in range(m):
            for j in range(n):
                if input[i][j] == "S": return (i, j, 1, 0)

    def BFS(start, input):
        explored = deque([start])
        visited = set()
        visited.add((start[0], start[1]))
        while explored:
            i, j, e, d = explored.popleft() # row, col, height, depth (steps)
            if input[i][j] == "E":
                return d
            for nei in [(i, j-1), (i, j+1), (i-1, j), (i+1, j)]:
                if nei[0] < 0 or nei[0] >= m or nei[1] < 0 or nei[1] >= n or nei in visited:
                    continue
                if input[nei] == "E":
                    height = elevationMap['z']
                else:
                    height = elevationMap[input[nei]]
                if height <= e + 1:
                    visited.add(nei)
                    explored.append((nei[0], nei[1], height, d + 1))
                
    start = getStartingPosition(input)

    return BFS(start, input)
```

```python
findShortestPath(mat)
```

```python
def findShortestPathP2(input):
    m, n = input.shape[0], input.shape[1]
    ele = string.ascii_lowercase # a-z
    elevationMap = {symbol: height for height, symbol in enumerate(ele, start = 1)}

    def getEndingPosition(input):
        for i in range(m):
            for j in range(n):
                if input[i][j] == "E": 
                    height = elevationMap['z']
                    return (i, j, height, 0)

    def BFS(start, input):
        '''To find optimal starting point, do BFS starting from the 
        end, exploring each neighbour as long as the height difference
        is at most 1
        '''
        explored = deque([start])
        visited = set()
        visited.add((start[0], start[1]))
        while explored:
            i, j, e, d = explored.popleft() # row, col, height, depth (steps)
            if input[i][j] == "a" or input[i][j] == "S":
                return d
            for nei in [(i, j-1), (i, j+1), (i-1, j), (i+1, j)]:
                if nei[0] < 0 or nei[0] >= m or nei[1] < 0 or nei[1] >= n or nei in visited:
                    continue
                if input[nei] == "S":
                    height = elevationMap['a']
                else:
                    height = elevationMap[input[nei]]
                if e - height <= 1:
                    visited.add(nei)
                    explored.append((nei[0], nei[1], height, d + 1))
    
    start = getEndingPosition(input)
    return BFS(start, input)
```
