```python
def findElfWithMaxCal(input):
    maxCal = 0
    for record in input:
        tot_calories = sum([int(n) for n in record.split("\n")])
        if tot_calories > maxCal:
            maxCal = tot_calories
    return maxCal
```

```python
def findTotalStratScore(input):
    # Opponent - A(Rock), B(Paper), C(Scissors)
    # Player - X(Rock), Y(Paper), Z(Scissors)
    player_scores = {
        "X": 1,
        "Y": 2,
        "Z": 3
    }
    '''Winning states:
    1) Paper beats Rock - (A Y)
    2) Rock beats scissors - (C X)
    3) Scissors beats paper - (B Z)
    Draw states: (A X), (B Y), (C Z)
    '''
    win_states = set([('A', 'Y'), ('C', 'X'), ('B', 'Z')])
    draw_states = set([('A', 'X'), ('B', 'Y'), ('C', 'Z')])
    score = 0
    for round in input:
        opp, player = round.rstrip().split(" ")
        if (opp, player) in win_states:
            score += (player_scores[player] + 6)
        elif (opp, player) in draw_states:
            score += (player_scores[player] + 3)
        else:
            score += player_scores[player]

    return score
```

```python
findTotalStratScore(strats)
```

```python
def findTotalStratScoreP2(input):
    # Opponent - A(Rock), B(Paper), C(Scissors)
    # Player(Outcome) - X(Lose), Y(Draw), Z(Win)
    player_scores = {
        "X": 1,
        "Y": 2,
        "Z": 3
    }
    win_states = set([('A', 'Y'), ('C', 'X'), ('B', 'Z')])
    draw_states = set([('A', 'X'), ('C', 'Z'), ('B', 'Y')])
    lose_states = set([('A', 'Z'), ('C', 'Y'), ('B', 'X')])
    score = 0
    for round in input:
        opp, state = round.rstrip().split(" ")
        if state == "Z":
            player = [x for x in win_states if x[0] == opp][0][1]
            score += (player_scores[player] + 6)
        elif state == "Y":
            player = [x for x in draw_states if x[0] == opp][0][1]
            score += (player_scores[player] + 3)
        else:
            player = [x for x in lose_states if x[0] == opp][0][1]
            score += player_scores[player]
    
    return score
```

```python
import string
def findTotalItemPriorities(input):
    letters = string.ascii_letters
    ans = 0
    # Build dict of priority values
    priorities = {letter: priority for priority, letter in enumerate(letters, start = 1)}
    for sack in input:
        items = sack.rstrip()
        mid = len(items) // 2
        compart_1, compart_2 = set(items[:mid]), set(items[mid:])
        overlaps = compart_1 & compart_2
        if overlaps:
            ans += sum([priorities[x] for x in overlaps])
    return ans
```

```python
findTotalItemPriorities(rucksacks)
```

```python
def findTotalItemPrioritiesP2(input):
    letters = string.ascii_letters
    ans = 0
    # Build dict of priority values
    priorities = {letter: priority for priority, letter in enumerate(letters, start = 1)}
    for i in range(0, len(input) - 2, 3):
        group = [set(x.rstrip()) for x in input[i:i+3]]
        badge_item = set.intersection(*group)
        ans += priorities[badge_item.pop()]

    return ans
```

```python
from operator import itemgetter
def findNumOverlappingAssignments(input):
    ans = 0
    for assignment in input:
        pair = assignment.rstrip().split(',')
        pair_lst = [tuple(map(int, pair[0].split('-'))), tuple(map(int, pair[1].split('-')))]
        pair_lst.sort(key = itemgetter(0))
        # Check if 2nd pair is fully contained in 1st pair
        if pair_lst[0][1] >= pair_lst[1][0] and pair_lst[1][1] <= pair_lst[0][1]:
            ans += 1
        # if starts are tied, check if 1st pair is fully contained in 2nd pair
        elif pair_lst[0][0] == pair_lst[1][0] and pair_lst[0][1] <= pair_lst[1][1]:
            ans += 1
    return ans
```

```python
findNumOverlappingAssignments(assignments)
```

```python
def findNumOverlappingAssignmentsP2(input):
    ans = 0
    for assignment in input:
        pair = assignment.rstrip().split(',')
        pair_lst = [tuple(map(int, pair[0].split('-'))), tuple(map(int, pair[1].split('-')))]
        pair_lst.sort(key = itemgetter(0))
        if pair_lst[0][1] >= pair_lst[1][0]:
            ans += 1
    return ans
```

```python
from collections import defaultdict
from itertools import groupby

def parser(input):
    '''Parse stack input
    '''
    stackMap = defaultdict(list)
    for line in reversed(input[:input.rfind("\n")].split("\n")):
        out = []
        for k,g in groupby(line.split(" "), lambda x: x == ''):
            if k: out.extend(['' for i in range(len(list(g)) // 4)])
            else: out.extend(g)
        # print(out)
        for idx, e in enumerate(out):
            if e: stackMap[idx + 1].append(e)
    return stackMap
```

```python
stackMap = parser(data[0])
for k,v in stackMap.items():
    print(k,v)
```

```python
def findCrateAssignments(moves, stacks):
    for line in moves.split("\n"):
        l = line.rstrip().split(" ")
        num, fr, to = int(l[1]), int(l[3]), int(l[-1])
        stacks[to].extend(reversed(stacks[fr][-num:])) # append new crates 
        stacks[fr] = stacks[fr][:len(stacks[fr]) - num] # remove appended crates
    out = []
    for v in stacks.values():
        out.append(v[-1][1:-1])
    return ''.join(out)
```

```python
class Directory:
    '''Object that represents a single directory
    '''
    def __init__(self, name, parent_dir):
        self.name = name
        self.parent = parent_dir
        self.subDirs = {}   # subdirs must be unique in any given dir
        self.files = []
        self.size = 0

    def add_subdir(self, subdir):
        self.subDirs[subdir] = Directory(subdir, self.name)

    def add_file(self, file, size):
        self.files.append((file, size))

    def compute_size(self):
        for file in self.files:
            self.size += int(file[1])
        
        for dir in self.subDirs.values():
            self.size += dir.compute_size()

        return self.size

    def __repr__(self):
        return f"({self.name}, {self.parent})"
```

```python
def findTotalSizeOfDirectories(input):
    root = Directory("/", None)
    path = [root]
    for line in input[2:]:
        l = line.rstrip().split(" ")
        prev = path[-1]
        if l[0] == "dir":
            prev.add_subdir(l[1])
            print(f"Added a sub-directory, {l[1]} into current directory: {prev.name}")
        elif l[1] == "cd" and l[2] != '..':
            path.append(prev.subDirs.get(l[2]))
            print(f"Cd into {l[2]} from current directory: {prev.name}")
        elif l[0].isnumeric():
            prev.add_file(l[1], l[0])
            print(f"Added a file, {l[1]} to current directory: {prev.name}")
        elif l[1] == "cd" and l[2] == '..':
            path.pop()
    
    root.compute_size()
    def helper(subdirs, ans):
        for dir in subdirs.values():
            if dir.size <= 100000:
                ans.append(dir.size)
            if len(dir.subDirs) > 0:
                helper(dir.subDirs, ans)
        return ans

    ans = helper(root.subDirs, [])
    return sum(ans)
```

```python
def findSmallestDirToDelete(input):
    root = Directory("/", None)
    path = [root]
    for line in input[2:]:
        l = line.rstrip().split(" ")
        prev = path[-1]
        if l[0] == "dir":
            prev.add_subdir(l[1])
        elif l[1] == "cd" and l[2] != '..':
            path.append(prev.subDirs.get(l[2]))
        elif l[0].isnumeric():
            prev.add_file(l[1], l[0])
        elif l[1] == "cd" and l[2] == '..':
            path.pop()
    
    root.compute_size()
    free_space = 70000000 - root.size
    min_space_to_free = 30000000 - free_space
    def helper(subdirs, ans):
        for dir in subdirs.values():
            if dir.size >= min_space_to_free:
                ans.append(dir.size)
            if len(dir.subDirs) > 0:
                helper(dir.subDirs, ans)
        return ans

    ans = helper(root.subDirs, [])
    ans.sort()
    return ans[0]
```

```python
def findNumVisibleTrees(input):
    num_visible = (input.shape[0] * 4) - 4  # count trees on edges as visible
    def is_visible(x, y):
        # Check along leftside of row of coordinate
        if input[x, :y].max() < input[x][y]:
            return True
        # Check along rightside of row of coordinate
        elif input[x, y+1:].max() < input[x][y]:
            return True
        # Check along top of column of coordinate
        elif input[:x, y].max() < input[x][y]:
            return True
        # Check along bottom of column of coordinate
        elif input[x+1:, y].max() < input[x][y]:
            return True
        else:
            return False

    for i in range(1, input.shape[0] - 1):
        for j in range(1, input.shape[1] - 1):
            if is_visible(i, j): num_visible += 1

    return num_visible
```

```python
def findMaxScenicScore(input):
    max_score = 0  
    def compute_scenic_score(x, y):
        left, right, top, bottom = 0, 0, 0, 0   # viewing distances
        # Check along leftside of row of coordinate
        for height in reversed(input[x, :y]):
            if height < input[x][y]: left += 1
            else: left += 1; break
        # Check along rightside of row of coordinate
        for height in input[x, y+1:]:
            if height < input[x][y]: right += 1
            else: right += 1; break
        # Check along top of column of coordinate
        for height in reversed(input[:x, y]):
            if height < input[x][y]: top += 1
            else: top += 1; break
        # Check along bottom of column of coordinate
        for height in input[x+1:, y]:
            if height < input[x][y]: bottom += 1
            else: bottom += 1; break
        return left*right*top*bottom

    for i in range(input.shape[0]):
        for j in range(input.shape[1]):
            scenic_score = compute_scenic_score(i, j)
            if scenic_score > max_score: max_score = scenic_score
    
    return max_score
```

```python
import math

def is_touching(head_pos, tail_pos):
    if math.dist(head_pos, tail_pos) >= 2:
        return False
    else:
        return True
    
def move_tail(head_pos, tail_pos):
    # Check if head is 2 steps right from tail
    if head_pos[1] == tail_pos[1] and head_pos[0] - tail_pos[0] > 0:
        tail_pos[0] = tail_pos[0] + 1
    # Check if head is 2 steps left from tail
    elif head_pos[1] == tail_pos[1] and head_pos[0] - tail_pos[0] < 0:
        tail_pos[0] = tail_pos[0] - 1
    # Check if head is 2 steps above tail
    elif head_pos[0] == tail_pos[0] and head_pos[1] - tail_pos[1] > 0:
        tail_pos[1] = tail_pos[1] + 1
    # Check if head is 2 steps below tail
    elif head_pos[0] == tail_pos[0] and head_pos[1] - tail_pos[1] < 0:
        tail_pos[1] = tail_pos[1] - 1
    # Head and tail are not on same row or column
    else:
        if head_pos[0] > tail_pos[0] and head_pos[1] > tail_pos[1]:
            # head is top diagonal right of tail
            tail_pos[0], tail_pos[1] = tail_pos[0] + 1, tail_pos[1] + 1
        elif head_pos[0] > tail_pos[0] and head_pos[1] < tail_pos[1]:
            # head is bottom diagonal right of tail
            tail_pos[0], tail_pos[1] = tail_pos[0] + 1, tail_pos[1] - 1
        elif head_pos[0] < tail_pos[0] and head_pos[1] > tail_pos[1]:
            # head is top diagonal left of tail
            tail_pos[0], tail_pos[1] = tail_pos[0] - 1, tail_pos[1] + 1
        elif head_pos[0] < tail_pos[0] and head_pos[1] < tail_pos[1]:
            # head is bottom diagonal left of tail
            tail_pos[0], tail_pos[1] = tail_pos[0] - 1, tail_pos[1] - 1

    return tail_pos
```

```python
def findNumPositionsVisited(input):
    head, tail = [0, 0], [0, 0] # set origin as initial position
    visited = set()
    visited.add((0, 0))
    for move in input:
        dir, steps = move.rstrip().split(" ")
        steps = int(steps)
        # print(f"Moving {steps} steps in {dir} direction...")
        while steps > 0:
            if dir == 'L':
                head[0] = head[0] - 1
            elif dir == "R":
                head[0] = head[0] + 1
            elif dir == "U":
                head[1] = head[1] + 1
            elif dir == "D":
                head[1] = head[1] - 1
            steps -= 1

            if not is_touching(head, tail):
                tail = move_tail(head, tail)
                if tuple(tail) not in visited: visited.add(tuple(tail)) 
        # print(f"Current position of H/T: {head}, {tail}")
    return len(visited)   
        
```

```python
from functools import reduce

def findNumPositionsVisitedP2(input):
    knots = [[0, 0] for i in range(10)]
    head = knots[0]
    visited = set()
    visited.add((0, 0))
    def simulate(head_pos, tail_pos):
        if not is_touching(head_pos, tail_pos):
            return move_tail(head_pos, tail_pos)
        else:
            return tail_pos

    for move in input:
        dir, steps = move.rstrip().split(" ")
        steps = int(steps)
        # print(f"Moving {steps} steps in {dir} direction...")
        while steps > 0:
            if dir == 'L':
                head[0] = head[0] - 1
            elif dir == "R":
                head[0] = head[0] + 1
            elif dir == "U":
                head[1] = head[1] + 1
            elif dir == "D":
                head[1] = head[1] - 1
            steps -= 1
            # Tail of each knot will be head for subsequent knot
            tail = reduce(lambda x, y: simulate(x, y), knots)
            if tuple(tail) not in visited: visited.add(tuple(tail))
        # print(f"Current position of H/T: {head}, {tail}")

    return len(visited)
```

```python
def findSignalStrengths(input):
    x, cycle = 1, 0
    signals = []
    for op in input:
        cycle += 1
        instr = op.rstrip().split(" ")
        if cycle % 40 == 20:
            signals.append(x * cycle)
            print(f"X: {x}, Cycle: {cycle}")
        if instr[0] == "addx":
            cycle += 1
            if cycle % 40 == 20:
                signals.append(x * cycle)
                print(f"X: {x}, Cycle: {cycle}")
            x += int(instr[1])
    
    return sum(signals)
```
