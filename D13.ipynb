```python
def findIndicesOfOrderedPackets(input):
    def isPacketOrdered(p1: list, p2: list, depth=1):
        '''Compare only the 1st element of each list i.e.
        [1, ...] and [2, ...] are always in order. If either
        one of the element is not a list, we convert to a list
        and redo the comparison i.e. [1,...] and [[2, 3],...] ->
        [[1],...] and [[2, 3],...]. If both integers are equal,
        continue comparison to the next element of each list.

        When the left list is exhausted first, they are in order. 
        If the right list is exhausted first, they are not in order.
        If they are both exhausted, continue comparison to the next 
        element. 
        '''
        while p1 and p2:
            e1, e2 = p1[0], p2[0]
            print("\t"*depth + f"Compare {e1} and {e2}")
            if isinstance(e1, int) and isinstance(e2, int):
                if e1 > e2: return False
                elif e1 < e2: return True
            else:
                if isinstance(e1, int): e1 = [e1]
                if isinstance(e2, int): e2 = [e2]
                result = isPacketOrdered(e1, e2, depth + 1)
                if result is not None: return result
            p1.pop(0)
            p2.pop(0)

        if p1: 
            return False # right list is exhausted first
        elif p2:
            return True # left list is exhausted first

    indexes = []
    for idx, pair in enumerate(input, start = 1):
        p1, p2 = pair.split("\n")
        p1, p2 = list(eval(p1)), list(eval(p2))
        print(f"Compare {p1} vs {p2}")
        if isPacketOrdered(p1, p2):
            print("Packet in order!") 
            indexes.append(idx)
        print("\n")

    return sum(indexes)
```

```python
findIndicesOfOrderedPackets(data)
```
